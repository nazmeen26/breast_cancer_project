```python
df_minority=df[df['OverallSurvivalStatus']==0]
df_majority=df[df['OverallSurvivalStatus']==1]
```

```python
from sklearn.utils import resample 
df_minority_upsampled=resample(df_minority, replace=True,
         n_samples=len(df_majority),
         random_state=42
         )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled['OverallSurvivalStatus'].value_counts()
```

```python
df_upsampled.corr()
```

```python
correlation_matrix = df_upsampled.corr()
plt.figure(figsize=(21, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',fmt=".2f") 
plt.title('Correlation Matrix')
plt.show()
```

```python
df_upsampled.to_csv('Breast Cancer METABRIC_final.csv',index=False)
```
